Select Income-Related Data

```python
inc = data.loc[:,['personid','date','thearn','thprpinc','thtrninc','thothinc','thtotinc']]
inc = inc.sort_values(by=['personid','date'])
inc.head(10)
```

Set Zero Income or Negative Income to Some Arbitrarily Small Value - For Log Differences

```python
# Replace Zero Values for All Data
inc = inc.replace(0, 0.01)
print(inc.head(10))

# Replace Negative Values for Property Income
inc['thprpinc'] = inc['thprpinc'].where(inc['thprpinc'] > 0, 0.01)
inc.head(10)
```

Calculate First-Differences

```python
import numpy as np

var_list = ['thearn','thprpinc','thtrninc','thothinc','thtotinc']

for i,N in enumerate(var_list):
    
    # Differences
    inc['d_' + N] = inc.groupby('personid')[N].diff()

    # Log-Differences
    inc['log_' + N] = np.log(inc[N])
    inc['dlog_' + N] = inc.groupby('personid')['log_' + N].diff()

inc.head(10)
```

Describe Adjusted Data

```python
inc.describe()
```

Drop Missing Values

```python
inc = inc.dropna()
inc.head(10)
```

Validate Total Income

```python
inc['thtotinc_2'] = inc['thearn'] + inc['thprpinc'] + inc['thtrninc'] + inc['thothinc']
```

Plot Histogram: Annual Income Changes

```python
import matplotlib.pyplot as plt

var_list = ['thearn','thprpinc','thtrninc','thothinc','thtotinc']
#var_list = ['thtotinc']

bins=100 # Number of bins in each histogram

for i,N in enumerate(var_list):
    
    fig,axis=plt.subplots(1,2,figsize=(20,5))
    
    x1 = inc['d_' + N]
    xx=axis[0].hist(x1,bins,density=True)
    
    x2 = inc['dlog_' + N]
    xx=axis[1].hist(x2,bins,density=True)
    
    print(N)
    plt.show()
```
